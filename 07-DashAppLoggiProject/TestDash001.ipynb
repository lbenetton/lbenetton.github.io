```python
@app.callback(
    Output('my-graph', 'figure'),
    [Input('agency', 'value'),
     Input('day', 'value')])
def update_graph(agency, day):
    #dff = df[df['ag'] == agency]
    dff = df[(df['ag'] == agency) & (df['td'] == day )]

    return {
        'data': [dict(
            x=dff['hour'],
            y=dff['drivers'],
            type='bar',
            marker=dict(color='rgb(80,90,109)')
            
        )],
        'layout': dict(
            title= 'Drivers Availability by Hour'),
            
        
    }
```
